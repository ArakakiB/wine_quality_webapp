Vou fazer um modelo beeem simples só para realizar o deploy

```python
import requests
import pandas as pd
import numpy  as np
import pickle
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model    import ElasticNet
from sklearn.metrics         import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing   import MinMaxScaler
```

```python
csv_url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
df_raw = pd.read_csv( csv_url, sep=';' )
```

```python
df1 = df_raw.copy()
```

como temos muitos dados dessa coluna que se concentram mais em um ponto (entre 0 e 30), vamos colocar essa coluna em uma escola de 0 a 1

```python
sns.displot(df1['free sulfur dioxide'], kde=False)
```

```python
mms = MinMaxScaler()
```

```python
# realizando o rescaler (colocar os numeros em escala de 0 a 1)
df1['free sulfur dioxide'] = mms.fit_transform(df1[['free sulfur dioxide']].values)
```

E para fazer o deploy, nos precisamos criar um arquivo para guardar essas transformacoes para os novos dados que chegarem.  
Para isso vamos criar um pickle que é como se fosse uma caixa em que guardamos as transformacoes (tipo uma variavel/classe)

```python
# guandando o scaler em um pickle
pickle.dump(mms, open('C:/Users/Araka/Dropbox/Estudos/Cursos/YT/Megairom/deploy/free_sulfur_dioxide_scaler.pkl', 'wb'))
```

```python
# perceba que a escala nao alterou os valores entre as variaveis
sns.displot(df1['free sulfur dioxide'], kde=False)
```

Vamos fazer para a outra coluna tambem, mas antes:

segundo o que o champs falou, um modelo de ml funciona melhor com uma distribuicao normal, e aqui nos temos uma distribuicao parecida com a de poisson (ta tudo de um lado so). Entao vamos transformar isso para uma distribuicao normal.

```python
sns.displot(df1['total sulfur dioxide'], kde=False)
```

```python
# transformando em uma distribuicao mais proxima da normal
df1['total sulfur dioxide'] = np.log1p(df1['total sulfur dioxide'])
```

Agora da pra ver que funfou, tambem mudou a escala

```python
sns.displot(df1['total sulfur dioxide'], kde=False)
```

Agora vamos fazer o rescaler para essa coluna

```python
df1['total sulfur dioxide'] = mms.fit_transform(df1[['total sulfur dioxide']].values)
```

```python
sns.displot(df1['total sulfur dioxide'], kde=False)
```

```python
# guandando o scaler em um pickle
pickle.dump(mms, open('C:/Users/Araka/Dropbox/Estudos/Cursos/YT/Megairom/deploy/total_sulfur_dioxide_scaler.pkl', 'wb'))
```

```python
train, test = train_test_split( df1 ) #75% e 25% test

# train and test dataset
x_train = train.drop( 'quality', axis=1 )
y_train = train['quality']

x_test = test.drop( 'quality', axis=1 )
y_test = test['quality']
```

```python
# model definition
model = ElasticNet( alpha=0.5, l1_ratio=0.5, random_state=42 )
```

```python
# training
model.fit( x_train, y_train )
```

```python
# prediction
pred = model.predict( x_test )
```

```python
rmse = np.sqrt( mean_squared_error( y_test, pred ))
mae = mean_absolute_error( y_test, pred )
r2  = r2_score( y_test, pred )
```

```python
print( 'RMSE: {}'.format( rmse ) )
print( 'MAE: {}'.format( mae ) )
print( 'R2: {}'.format( r2 ))
```
